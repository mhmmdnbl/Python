```python
# Data Visualization
# Seaborn is a  statistical data visualization library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
# import warnings
# warnings.filterwarnings('ignore')
```

```python
# built-in datasets of seaborn
sns.get_dataset_names()
```

```python
dataset =sns.load_dataset('iris')
dataset
```

```python
# lineplot() used to generate a line graph
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Line Graph Using Seaborn')
plt.show()
```

```python
# adding style to seaborn
sns.set_style('whitegrid',{'grid.color':'k','grid.linestyle':':'})
sns.lineplot(x='sepal_length',y='sepal_width',data=dataset)
plt.title('Line Graph Using Seaborn')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
# scatter plot
# scatterplot()

sns.scatterplot(x='Attack',y='Defense',data=pokemon)
plt.title('Scatter Plot Using Seaborn')
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
colors = ['r','']
sns.scatterplot(x='Attack',y='Defense',hue='Generation',data=pokemon)
plt.title('Scatter Plot Using Seaborn')
plt.show()
```

```python
pkmn_type_colors = ['#3ae40c',  # Grass
                    '#e52106',  # Fire
                    '#0AE3D6',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#ECF709',  # Dragon
                    '#7A786C',  # Dark
                    '#D3D1C0',  # Steel
                    '#91E8EE',  # Flying
                   ]
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Type 1',hue='Type 1',palette=pkmn_type_colors,data=pokemon)
plt.title('Count Plot Using Seaborn')
plt.xticks(rotation=45)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# who provides tips more [Male of Female]
# col = column variable
# split your dataset by the unique value and make a seperate plot for each

sns.relplot(x='total_bill',y='tip',col='sex',hue='smoker',data=tips)
plt.show()
```

```python
# on which day waiters are getting more tips

sns.relplot(x='total_bill',y='tip',col='time',hue='day',data=tips)
plt.show()
```
