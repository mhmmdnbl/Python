```python
# Data Visualization
# Matplotlib
# https://matplotlib.org/

# Data visualization is the representation of data through use of common graphics,
# such as charts, plots, infographics and even animations.
# The primary goal of data visualization is to make data more accessible and
# easier to interpret, allowing users to identify patterns, trends, and outliers quickly.
# Matplotlib is a comprehensive library for creating static, animated,
# and interactive visualizations in Python.
# Matplotlib makes easy things easy and hard things possible.
# You can generate plots, histograms, power spectra, bar charts, errorcharts,
# scatterplots, etc., with just a few lines of code.
# The matplotlib library contains several modules. The matplotlib.pyplot module is
# designed to create colorful plots.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import  random
```

```python
# Line Graph
# draw a line from point to point
# plot (x,y)

x = random.randint(10,size=(6))   # will generate 6 random coordinates for x-axis from 1-10
print(x)
y = random.randint(5,size=(6))    # will generate 6 random coordinates for y-axis from 1-5
print(y)
plt.plot(x,y)                     # used to display the graph
```

```python
x = [5,10,15]
y = [3,11,5]
plt.plot(x,y)
plt.title('My First Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Adding styles to a graph
# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html

from matplotlib import style
# style.available
style.use('ggplot')
x1 = [5,8,10]
y1 = [12,16,6]
x2 = [6,9,11]
y2 = [7,15,7]
x3 = [5,6,11]
y3 = [8,8,8]
plt.plot(x1,y1,color='green',linewidth=3,linestyle='solid',label='Product 1')     # linestyle = 'solid' / 'dashed' / 'dotted'
plt.plot(x2,y2,color='yellow',linewidth=3,linestyle='solid',label='Product 2')
# plt.plot(x3,y3,color='red',linewidth=3,linestyle='dashed',label='KPI')
plt.title('Graph Using Style')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=2)   # use to display label on graph, and can also change the location of labels
plt.grid(True,color='darkblue')   # k --> black color
plt.show()
```

```python
# Bar Graph
# bar()

plt.bar([1,3,5,7,9],[5,2,7,8,2],label='Data1')
plt.bar([2,4,6,8,10],[8,6,2,5,6],label='Data2')
plt.title('Bar Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
# Horizontal Bar Graph
# barh()

products = ['Computers','Tablets','Mouse','Keyboard','Laptop']
quantities = [320,670,280,200,500]
plt.barh(products,quantities)
plt.title('Stock Inventory')
plt.xlabel('Quantity')
plt.ylabel('Products')
plt.show()
plt.savfig("myBarGraph.jpg")
```

```python
# Histogram
# a histogram is used to show the distribution of data
# for example : if you have the ages of 100 people , a histogram help you see how many people are in their in 20's
# 30s , 40s
# ages = [18,21,22,25,26,30,34,35,36,40,42,45,46,50]
# bins = 5
# divide the range of ages (18-50) into 5 equal sized groups
# 18-24 25-31 32-38 39-45 46-52

ages = [18,21,22,25,26,30,34,35,36,40,42,45,46,50]
plt.hist(ages,bins=5,color='skyblue',edgecolor='black')
plt.title('My First Histogram')
plt.xlabel('Age')
plt.ylabel('No. of People')
plt.show()
```

```python
# Subplots
# It allows us to create multiple plots within a single figure

# Plot 1
x1 = random.randint(10,size=(6))
y1 = random.randint(6,size=(6))
plt.subplot(2,3,1)        # (a,b,c) ----> a = number of row, b = number of column, c = position of graph
plt.plot(x1,y1,color='blue',label='Plot 1')
plt.title('Graph 1')
plt.legend()

# Plot 2
x2 = random.randint(10,size=(6))
y2 = random.randint(6,size=(6))
plt.subplot(2,3,2)
plt.plot(x2,y2,color='red',label='Plot 2')
plt.title('Graph 2')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12, 6))

colors = ['green', 'red', 'blue', 'purple', 'orange', 'brown']

x_max = [10, 20, 15, 12, 18, 25]
y_max = [6, 10, 8, 9, 7, 12]

# Loop --> 6 subplots in 2x3 layout
for i in range(6):
    x = np.random.randint(x_max[i], size=(7))
    y = np.random.randint(y_max[i], size=(7))
    plt.subplot(2, 3, i + 1)  # 2 rows, 3 columns, position i+1
    plt.plot(x, y, color=colors[i])
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title(f'Plot {i + 1}')

# Adjust layout and show all subplots instead of plt.subplots_adjust()
plt.tight_layout()
plt.show()
```

```python
# Introduction To
# The Figure and Axes Concept

# a matplotlib chart is comprises of 2 things
#    a.) fig : the container (whole window or page)
#    b.) ax  : the plot inside it
```

```python
plt.figure(facecolor='blue')    # add a background color
plt.plot([1,2,3,4],[10,20,25,40])
ax = plt.gca()                      # get current grid axis
ax.set_facecolor('lightblue')
plt.title('Without fig,ax')
plt.show()
```

```python
fig,ax = plt.subplots()
fig.set_facecolor('blue')
plt.plot([1,2,3,4],[10,20,25,40])
ax.set_facecolor('lightblue')
plt.title('With fig,ax')
ax.tick_params(axis='x',colors='white')
ax.tick_params(axis='y',colors='white')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3])

# Get the list of x-tick label objects
x_labels = ax.get_xticklabels()

# Set the color of the first x-tick label to red
x_labels[0].set_color("red")

# Get the list of y-tick label objects
y_labels = ax.get_yticklabels()

# Set the color of the last y-tick label to blue
y_labels[-1].set_color("blue")

plt.show()
```
